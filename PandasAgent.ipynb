````python
# Read pandas file

import pandas as pd
df = pd.read_csv('hoanghamobile.csv')
````

````python
df
````

````python
# convert the current price in the collum current_price to float, if Nan -> convert to 0
# '1,590,000 ₫' -> 1590000
# Giá: Liên hệ -> 0

df = df.dropna(subset=['current_price'])
df['current_price'] = df['current_price'].str.replace('₫', '').replace('Giá: Liên hệ', '0')
df['current_price'] = df['current_price'].str.replace(',', '')
df['current_price'] = df['current_price'].astype(float)
df
````

````python
# write query to extract which row has current_price greater than 4000000

query = df.query('current_price > 4000000')
query
````

````python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4")
````

````python
from langgraph.graph import START, StateGraph
from langgraph.prebuilt import tools_condition
from langgraph.prebuilt import ToolNode
from langchain.tools import tool
from IPython.display import Image, display
from pydantic import BaseModel
import pandas as pd

# Initialize LLM
llm = ChatOpenAI(model="gpt-4", temperature=0)

# Define tools using the @tool decorator
def pandas_query_tool(query: str) -> str:
    """
    Executes a pandas query string on the DataFrame.
    """
    try:
        # Use eval to execute the pandas query
        result = df.query(query)
        return result.to_string(index=False)
    except Exception as e:
        return f"Error executing query: {str(e)}"


# Bind tools to LLM
tools = [pandas_query_tool]
llm_with_tools = llm.bind_tools(tools)

````

````python
from langchain_core.messages import SystemMessage

def reasoner(state: MessagesState):
    """
    Reasoner node: Processes the user's input and determines the next action (tool call or direct response).
    """
    # Construct a system message to guide the LLM in producing queries or function calls
    sys_msg_content = (
        "You are a data assistant. Your role is to help users query a pandas DataFrame. "
        "The DataFrame contains a column named 'current_price' which holds numerical values. "
        "When users provide a query, interpret it and generate a pandas query string "
        "that operates specifically on the 'current_price' column. "
        "Always return your response in the following format:\n\n"
        "```\n"
        "{\n"
        "  \"query\": \"<pandas_query_string>\"\n"
        "}\n"
        "```\n"
        "If the query cannot be formed, set `<pandas_query_string>` to an empty string (\"\").\n"
        "If the query requires execution, return it as a tool call using:\n"
        "`functions.pandas_query_tool({\"query\": \"<pandas_query_string>\"})`."
    )
    sys_msg = SystemMessage(content=sys_msg_content)

    # Invoke the LLM with the system message plus the user messages from state
    # The state should contain user messages. For example, if state["messages"]
    # is a list of HumanMessage or similar, we combine it with sys_msg.
    return {
        "messages": [
            llm_with_tools.invoke([sys_msg] + state["messages"])
        ]
    }
````

````python
# Graph builder
builder = StateGraph(MessagesState)  # Pass the schema class instead of memory

# Add nodes
builder.add_node("reasoner", reasoner)  # Reasoner node processes user queries
builder.add_node("tools", ToolNode(tools))  # ToolNode executes generated commands

# Add edges
builder.add_edge(START, "reasoner")  # Start from reasoner
builder.add_conditional_edges(
    "reasoner",
    tools_condition,  # Check if the result is a tool call
)
builder.add_edge("tools", "reasoner")  # Loop back to reasoner after tools execution

# Compile graph
react_graph = builder.compile()

# Display graph
display(Image(react_graph.get_graph(xray=True).draw_mermaid_png()))
````
